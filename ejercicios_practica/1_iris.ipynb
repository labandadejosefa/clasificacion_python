```python
import os
import platform

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from  sklearn import  datasets
```

# Recolectar datos
<div align="center"><img src="https://raw.githubusercontent.com/InoveAlumnos/dataset_analytics_python/master/images/Pipeline1.png" width="1000" align="middle"></div>

```python
iris=datasets.load_iris()
df_data = pd.DataFrame(data= iris.data, columns= iris.feature_names)
df_target = pd.DataFrame(data= iris.target, columns= ['species'])
print(df_data.head())
print(df_target.head())
```

```python
# Generar una copia de df_target llamada "df_target_categorical" en el cual se almacenará los tipos de iris como variables categóricas (texto)
df_target_categorical = df_target.copy()
```

# Procesar datos
<div align="center"><img src="https://raw.githubusercontent.com/InoveAlumnos/dataset_analytics_python/master/images/Pipeline2.png" width="1000" align="middle"></div>

```python
# Alumno: Utilizar apply y lambda expresion sobre df_target para modificar la columna "species" de df_target_categorical
# la lambda debe devolver:
# setosa --> si species = 0
# versicolor --> si species = 1
# virginica --> si species = 2

# Tip:
# df_target_categorical['species'] = df_target.apply(lambda...... 
df_target_categorical['species'] = df_target.apply(lambda x: 'setosa' if x.species == 0 else 'versicolor' if x.species == 1 else 'virginica', axis=1)
```

```python
# Alumno: Observar las primeras filas del nuevo df_target_categorical para corroborar que haya sido correcto el cambio
df_target_categorical.head()
```

```python
# Alumno: Contar cuantas especies de iris hay de cada una en el dataset df_target_categorical
df_target_categorical['species'].value_counts()
```

```python
# Concatenar ambos dataset
df = pd.concat([df_data, df_target_categorical], axis= 1)
df.head()
```

```python
# Alumno: Analice si hay elementos incompletos en el dataset "df_data" o "df_target"
des = df.describe()
des.loc['Nan'] = df.isna().sum()
des.loc['%Nan'] = (df.isna().mean())*100
des
```

```python
# Alumno: ¿Cuántas filas/datos hay para analizar?
print('Datos disponibles para analizar: ', df.shape[0])
```

# Entrenar modelo
<div align="center"><img src="https://raw.githubusercontent.com/InoveAlumnos/dataset_analytics_python/master/images/Pipeline4.png" width="1000" align="middle"></div>

```python
# Creamos el modelo base
class RandomBaseModel():
    def __init__(self):
        self.classes_ = [0, 1]
    def fit(self,X, y):
        self.classes_ = np.unique(y)
        return None

    def predict(self,X):
        rand = np.random.randint(0, len(self.classes_), size=X.shape[0])
        rand_clases = [self.classes_[x] for x in rand]
        return np.asarray(rand_clases)
```

```python
# Almuno: Crear un numpy array "X" con los features de entrada (todos excepto species)
# Crear un numpy array "Y" de la columna species


X = df.drop('species', axis=1).values #en el array X, todos MENOS "species"
y = df['species'].values 
```

```python
# Alumno: Cree los dataset de entrenamiento (train) y evaluación (test)
# utilizando la herramienta train_test_split de scikit-learn, utilice la proporción
# 70%30%

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42);
```

```python
# Almuno: Utilizar el modelo base y entrenarlo con el set de datos de train
# Luego obtener y_hay_base con las predicción del set de datos de test

#utilizo el modelo creado como Class, en el bloque más arriba. 1ero creo un objeto de esa clase:
random_model = RandomBaseModel();

# lo entreno
random_model.fit(X_train, y_train); 

#guardo los resultados de las predicciones como se pide
y_hat_base = random_model.predict(X_test);
```

```python
# Alumno: Entrenar un clasificador utilizando algunos de los siguientes algoritmos
# - Random forest
# - KNN
# - SVM

# importo las clases a utilizar para Random Forest & KNN
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

#1. Random Forest
# Creo el objeto de la clase RandomForestClassifier(). Elijo n_estimators & max_depth arbitrariamente para probar (posibilidad de ajustar al validar el modelo)
clf_forest = RandomForestClassifier(n_estimators=3, max_depth=4, criterion='entropy', random_state=0)

# lo entreno
clf_forest.fit(X_train, y_train)

# Guardo los resultados de las predicciones
y_hat_forest = clf_forest.predict(X_test)

#================================================

#2. KNN (K nearest neighbors)
# A diferencia del ejemplo_clase nº2 (frutas), todos los datos medidos en cm parecen guardar una proximidad entre sí, tal que permite evitar su normalización (Profes: ¡corregir, si no!).

# Creo -y entreno- el objeto de la clase KNeighborsclassifier() con n=3, pues sabemos que son 3 especies (setosa, versicolor, virginica).
clf_knn = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)

# Guardo los resultados de las predicciones
y_hat_knn = clf_knn.predict(X_test)

```

# Validar modelo
<img src="https://raw.githubusercontent.com/InoveAlumnos/dataset_analytics_python/master/images/Pipeline5.png" width="1000" align="middle">

```python
# Alumno: Calcular la exactitud (accuracy) del modelo base

# Accuracy del modelo creado con la class RandomBaseModel()
from sklearn.metrics import accuracy_score

print(f'Exactitud del modelo Base: {round(accuracy_score(y_test, y_hat_base, normalize=True),3)}')
```

```python
# Alumno: Calcular la exactitud (accuracy) del clasificador

#1. Accuracy del modelo de Random Forest
print(f'Exactitud del modelo Random Forest: {accuracy_score(y_test, y_hat_forest, normalize=True) }')
```

```python
#2. Accuracy del modelo de KNN
print(f'Exactitud del modelo KNN: {accuracy_score(y_test, y_hat_knn, normalize=True)}')
```

**Sobre el cálculo de accuracies**: i) El modelo RandomBaseModel, arroja una exactitud de tan sólo 0.31; ii) los modelos Random Forest y KNN, obtienen una exactitud igual a 1.0, siendo los valores más altos posibles, en precisión.

La elección inicial de los parámetros n_estimators=3 & max_depth=4 para el modelo KNN, resultó ser muy eficiente.

```python
# Dibujar la matriz de confusión con y_test e y_hat_base. Colocar los "labels" (titulos) correspondientes

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

#1. Modelo Base

# Creo la matriz -objeto- con los 2 valores a comparar
cm_base = confusion_matrix(y_test, y_hat_base)

# Creo la variable que almacenará la representación gráfica de la matriz
cmd_base = ConfusionMatrixDisplay(cm_base, display_labels=random_model.classes_)

# Especifico mapa de colores a utilizar (cmap=plt.cm.<COLOR>)
cmd_base.plot(cmap=plt.cm.Reds)

# Muestro la figura (¡Y elimino el encabezado "<sklearn.metrics._plot.confusion_matrix.ConfusionMatrixDisplay at 0x7f9e23f17760>" del output!)
plt.show()
```

```python
#2. Modelo Random Forest

# Creo la matriz -objeto- con los 2 valores a comparar
cm_forest = confusion_matrix(y_test, y_hat_forest)

# Creo la variable que almacenará la representación gráfica de la matriz
cmd_forest = ConfusionMatrixDisplay(cm_forest, display_labels=clf_forest.classes_)

# Especifico mapa de colores a utilizar (cmap=plt.cm.<COLOR>)
cmd_forest.plot(cmap=plt.cm.Accent_r)

# Muestro la figura
plt.show()
```

```python
#2. Modelo K Nearest Neighbors

# Creo la matriz -objeto- con los 2 valores a comparar
cm_knn = confusion_matrix(y_test, y_hat_knn)

# Creo la variable que almacenará la representación gráfica de la matriz
cmd_knn = ConfusionMatrixDisplay(cm_knn, display_labels=clf_knn.classes_)

# Especifico mapa de colores a utilizar (cmap=plt.cm.<COLOR>)
cmd_knn.plot(cmap=plt.cm.terrain_r)

# Muestro la figura
plt.show()
```
